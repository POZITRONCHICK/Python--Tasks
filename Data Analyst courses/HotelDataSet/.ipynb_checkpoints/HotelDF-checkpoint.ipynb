```python
import pandas as pd
import numpy as np
```

###  Считываем данные и смотрит шапку таблицы:

```python
hoteldf = pd.read_csv('C:/Users/POZI/Desktop/портфолио/Projects/HotelDataSet/bookings.csv', delimiter=';')
hoteldf.head(10)
```

### Смотрим размерность, название столбцов и тип данных в таблице:

```python
hoteldf.shape, hoteldf.info(), hoteldf.columns
```

### Переименовываем неудобные названия столбцов:

```python
hoteldf = hoteldf.rename(columns={'Hotel': 'hotel', 
                        'Is Canceled': 'is_canceled', 
                        'Lead Time': 'lead_time', 
                        'arrival full date': 'arrival_full_date',
                        'Arrival Date Year': 'arrival_date_year', 
                        'Arrival Date Month': 'arrival_date_month', 
                        'Arrival Date Week Number': 'arrival_date_week_number',
                        'Arrival Date Day of Month': 'arrival_date_day_of_month', 
                        'Stays in Weekend nights': 'stays_in_weekend_nights',
                        'Stays in week nights': 'stays_in_week_nights', 
                        'stays total nights': 'stays_total_nights', 
                        'Adults': 'adults', 
                        'Children': 'children',
                        'Babies': 'babies', 
                        'Meal': 'meal', 
                        'Country': 'country', 
                        'Reserved Room Type': 'reserved_room_type', 
                        'Assigned room type': 'assigned_room_type',
                        'customer type': 'customer_type', 
                        'Reservation Status': 'reservation_status', 
                        'Reservation status_date': 'reservation_status_date'})
hoteldf.head()
```

### Топ 5 стран по количеству успешных бронирований:

```python
hoteldf \
    .query('is_canceled == 0') \
    .groupby('country') \
    .agg({'is_canceled' : 'count'}) \
    .sort_values(by='is_canceled', ascending=False) \
    .head(5)
```

### На сколько ночей в среднем бронируют отели разных типов:

```python
hoteldf \
    .groupby('hotel') \
    .agg({'stays_total_nights' : 'mean'}) \
    .round(2)
```

### Иногда тип номера, полученного клиентом (assigned_room_type), отличается от изначально забронированного (reserved_room_type).
### Сколько подобных наблюдений встретилось в датасете?

```python
hoteldf \
    .query('reserved_room_type != assigned_room_type') \
    .value_counts('reserved_room_type') \
    .sum()
```

### На какой месяц чаще всего успешно оформляли бронь в 2016? Изменился ли самый популярный месяц в 2017?

```python
hoteldf \
    .query('arrival_date_year == 2016') \
    .query('is_canceled == 0') \
    .groupby('arrival_date_month') \
    .agg({'is_canceled' : 'count'}) \
    .sort_values(by='is_canceled', ascending=False) \
    .head(1)
```

```python
hoteldf \
    .query('arrival_date_year == 2017') \
    .query('is_canceled == 0') \
    .groupby('arrival_date_month') \
    .agg({'is_canceled' : 'count'}) \
    .sort_values(by='is_canceled', ascending=False) \
    .head(1)
```

### Сгруппируйте данные по годам и проверьте, на какой месяц бронирования отеля типа City Hotel отменялись чаще всего в каждый из периодов.

```python
hoteldf \
    .loc[hoteldf['hotel'] == 'City Hotel'] \
    .query('is_canceled == 1') \
    .groupby(['arrival_date_year', 'arrival_date_month']) \
    .agg({'is_canceled' : 'count'}) 
```

### Посмотрите на числовые характеристики трёх колонок: adults, children и babies. Какая из них имеет наибольшее среднее значение?

```python
hoteldf[['adults', 'children', 'babies']].mean()
```

### Создайте колонку total_kids, объединив столбцы children и babies. Для отелей какого типа среднее значение переменной оказалось наибольшим?

```python
hoteldf['total_kids'] = hoteldf.children + hoteldf.babies

hoteldf.groupby(['hotel']).agg({'total_kids' : 'mean'}).round(2)
```

### Не все бронирования завершились успешно (is_canceled), поэтому можно посчитать, сколько клиентов было потеряно в процессе. Иными словами, посчитать метрику под названием Churn Rate.

```python
hoteldf['has_kids'] = hoteldf.total_kids > 0
```

```python
round(hoteldf.query('is_canceled == 1 and has_kids == False').shape[0] / hoteldf.query('has_kids == False').shape[0] * 100, 2)
```

```python
round(hoteldf.query('is_canceled == 1 and has_kids == True').shape[0] / hoteldf.query('has_kids == True').shape[0] * 100, 2)
```
